```python
import matplotlib.pyplot as plt
```

```python
# This will display figure in the notebook wont pop a window to show the plot
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
# Functional way of doing
plt.plot(x,y,'r')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# OO object oriented

fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('x lable')
axes.set_ylabel('y lable')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.15,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('OUTER PLOT')

axes2.plot(y,x)
axes2.set_title('INNER PLOT')
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)

plt.tight_layout()

#axes is a list of axes objects

axes[0].plot(x,y,'r')
axes[0].set_title('FIRST PLOT')
axes[1].plot(y,x,'b')
axes[1].set_title('SECOND PLOT')

# for cur_axes in axes:
#     cur_axes.plot(x,y)
```

# Figure size and DPI

```python
#for a single fugure you need to create fig then axes explicitly
fig = plt.figure(figsize=(5,2),dpi=100)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
# for subplots you can get fig and axes automatically like below
fig, axes = plt.subplots(nrows=2,ncols=1,figsize=(4,4),dpi=80)

axes[0].plot(x,y,'r')
axes[1].plot(x,y,'b')

plt.tight_layout()
```

```python
fig.savefig('./output/my_matplot_pic.jpg')
```

```python
# with the same figure you can actually show two plots

fig = plt.figure(dpi=80)

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label = 'X Square', color = 'red', linestyle='dashed',linewidth=1, markersize=12)
ax.plot(x, x**3, label = 'X Cube')
ax.legend()
```

# PLOT Appearance

```python
fig = plt.figure(figsize=(4,4),dpi=80)

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='green', marker='o', linestyle='dotted', linewidth=2, markersize=5,  alpha = 0.5, animated=True)

# you can set x limit and y axes limits

# ax.set_xlim([0,1])
# ax.set_ylim([0,2])
```
