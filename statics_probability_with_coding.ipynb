```python
def get_categorical_value_counts(dataframe):
    df = dataframe
    cat_col = df.select_dtypes(include='object').columns
    for i in cat_col:
        print(i, '\n')
        print(df[i].value_counts())
        print("=="*15)
    return "**"*15
```

```python
get_categorical_value_counts(df)
```
